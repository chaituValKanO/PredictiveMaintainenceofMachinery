## Feature Engineering

### Component Replacement Log 

```python
componentsReplacementLog = componentsReplacementLog.sort_values(by='Comp_Replace_DateTime',ascending=True)
componentsReplacementLog.head(5)
```

#### Days btw replacing of diff components

```python
componentsReplacementLog['Comp_Timelapse_diff_replace'] = round(componentsReplacementLog.groupby('MachineID')['Comp_Replace_DateTime'].diff().dt.total_seconds()/86400, 2)
```

```python
componentsReplacementLog.Comp_Timelapse_diff_replace = componentsReplacementLog.Comp_Timelapse_diff_replace.fillna(0)
```

```python
#componentsReplacementLog['Comp_Timelapse_diff_replace'].unique()
```

```python
componentsReplacementLog.head(5)
```

#### Days btw replacing of same component

```python
componentsReplacementLog['Comp_Timelapse_same_replace'] = round(componentsReplacementLog.groupby(['MachineID', 'Comp_Replace_ComponentReplacement'])['Comp_Replace_DateTime'].diff().dt.total_seconds()/86400, 2)
```

```python
componentsReplacementLog.Comp_Timelapse_same_replace = componentsReplacementLog.Comp_Timelapse_same_replace.fillna(0)
```

```python
#componentsReplacementLog['Comp_Timelapse_same_replace'].unique()
```

```python
componentsReplacementLog.head(5)
```

```python
components_log_same_replace= round(componentsReplacementLog.groupby(['MachineID', 'Comp_Replace_ComponentReplacement'], as_index=False)['Comp_Timelapse_same_replace'].mean(), 2)
```

```python
components_log_same_replace.head(10)
```

```python
components_log_same_replace_pivot = pd.pivot_table(components_log_same_replace, index=["MachineID", "Comp_Replace_ComponentReplacement"], aggfunc='mean')
```

```python
components_log_same_replace_pivot
```

```python
components_log_same_replace_pivot = components_log_same_replace_pivot.unstack()
```

```python
components_log_same_replace_pivot = components_log_same_replace_pivot.fillna(0)
```

```python
components_log_same_replace_pivot
```

```python
#components_log_same_replace_pivot.to_csv('components_log_same_replace_pivot.csv')
```

#### Getting most recent replaced Component for each machine

```python
g1 = componentsReplacementLog.groupby('MachineID').agg(lambda x: x.iloc[-1]).unstack()['Comp_Replace_ComponentReplacement']
df = g1.reset_index()
df.columns = ['MachineID', 'RecentReplacedComp']
df.head(6)
```

#### Getting most replaced component for a machine

```python
s_value= componentsReplacementLog.groupby(['MachineID'])['Comp_Replace_ComponentReplacement'].value_counts()
s_res = s_value.groupby(level=[0]).nlargest(1).reset_index(level=[0], drop=True)

#print(s_res.index[0])
most_replaced = pd.DataFrame(list(s_res.index), columns = ['MachineID', 'MostReplacedComp'])
most_replaced.head(5)
```

#### Getting days for replacing btw diff components and same components

```python
components_log_diff_replace_pivot= round(componentsReplacementLog.groupby(['MachineID'], as_index=False)['Comp_Timelapse_diff_replace'].mean(), 2)
```

```python
components_log_diff_replace_pivot.head(5)
```

```python
components_log_same_replace_csv = pd.read_csv('C:\\Users\\chait\\Documents\\Insofe\\PhD\\scripts\\components_log_same_replace_pivot.csv')
components_log_same_replace_csv.head(5)
```

```python
components_log_replace_int1 = pd.merge(components_log_same_replace_csv, components_log_diff_replace_pivot, on='MachineID', how='outer')
components_log_replace_int1.head(5)
```

```python
components_log_replace_int2 = pd.merge(components_log_replace_int1, df, on='MachineID', how='outer')
components_log_replace_int2.head(5)
```

```python
components_log_replace_final = pd.merge(components_log_replace_int2, most_replaced, on='MachineID', how='outer')
components_log_replace_final.head(5)
```

```python
components_log_replace_final.columns = ['MachineID', 'comp_1_replace_avg', 'comp_2_replace_avg', 'comp_3_replace_avg', 'comp_4_replace_avg', 'Comp_Timelapse_diff_replace', 'RecentReplacedComp', 'MostReplacedComp']
```

```python
components_log_replace_final.isnull().sum(axis=0)
```
